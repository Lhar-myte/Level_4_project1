```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen, Request
import numpy as np
import pandas as pd
# empty list to store the scrapp data
Title = []
Price = []
Availability = []
Rating = []
Category = []
page_numbers = []
all_pages = []

for page_num in range(1,21):
    print(f"scrapping page{page_num}...")
    
    url = f"http://books.toscrape.com/catalogue/page-{page_num}.html"
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    page = urlopen(req)
    page_soup = soup(page, "html.parser")
    all_pages.append(page_soup)

    books = page_soup.find_all("article", class_="product_pod")
#     print(f"Found {len(books)} books on page {page_num}")
    
# add headers
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    page = urlopen(req)
  
   
    for book in books:
            title = book.h3.a["title"]
            price = book.find("p", class_="price_color").text
            availability = book.find("p", class_="instock availability").text.strip()
            rating = book.p["class"][1]

            Title.append(title)
            Price.append(price)
            Availability.append(availability)
            Rating.append(rating)
            Category.append("Books")
            page_numbers.append(page_num)
```

```python
df = pd.DataFrame({
    "Title": Title,
    "Price": Price,
    "Availability": Availability,
    "Rating": Rating,
    "Category": Category,
    "Page": page_numbers
})
df.info()
print(df)
df.to_csv("mystery.csv", index=False)
```
